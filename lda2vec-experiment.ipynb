```python
import pandas as pd
import os
```

```python
DATA_DIR = './tests/twenty_newsgroups/data'
data_file = os.listdir(DATA_DIR)[0]
df = pd.read_csv(os.path.join(DATA_DIR, data_file), sep="\t")
df.head()
```

```python
df.shape
```

```python
texts = df.texts.values.tolist()
```

### Ad-hoc text preprocessing 

```python
# In case you want to restart the process
# !rm -Rf ./data
```

```python
bad = set(["ax>", '`@("', '---', '===', '^^^', "AX>", "GIZ"])
```

```python
utils.run_preprocessing(texts, './', 'data', bad=bad, max_length=10000)
```

### Model training

```python
NUM_TOPICS = 20
EPOCHS = 20
```

```python
(idx_to_word, word_to_idx, freqs, embed_matrix, pivot_ids,
 target_ids, doc_ids, num_docs, vocab_size, embed_size) = utils.load_preprocessed_data('./', 'data')
```

** You can jump straight to [Topic Visualization](#Topic-visualization) section if you already trained your model **

```python
# for i in tf.get_collection(tf.GraphKeys.GLOBAL_VARIABLES):
#     print(i)
```

```python
tf.reset_default_graph()
m = model(num_docs,
          vocab_size,
          num_topics=NUM_TOPICS,
          embedding_size=embed_size,
          load_embeds=True,
          pretrained_embeddings=embed_matrix,
          freqs=freqs)
```

```python
m.train(pivot_ids, target_ids, doc_ids, len(pivot_ids), EPOCHS, idx_to_word=idx_to_word, switch_loss_epoch=5)
```

```python
!mkdir results
```

```python
m.save_weights_to_file(word_embed_path='./results/w_embed', 
                       doc_embed_path='./results/doc_embed', 
                       topic_embed_path='./results/topic_embed')
```
